```python
#importing required libraries
from keras.preprocessing import image
from keras.models import load_model
import numpy as np
from google.colab import files

#uploading the model
print("Upload your model.h5 file:")
uploaded_model = files.upload()

#uploading testing files
print("Upload your image files:")
uploaded_images = files.upload()

#loading the model
model = load_model('model.h5')
print("Model Loaded Successfully")

#defining the classify function
def classify(img_file):
    test_image = image.load_img(img_file, target_size=(256, 256), color_mode='grayscale')
    test_image = image.img_to_array(test_image)
    test_image = np.expand_dims(test_image, axis=0)
    result = model.predict(test_image)
    arr = np.array(result[0])
    print(arr)
    max_prob = arr.argmax()
    classes = ["NONE", "ONE", "TWO", "THREE", "FOUR", "FIVE"]
    prediction = classes[max_prob]
    print(img_file, prediction)

#printing all the classified functions
for img_file in uploaded_images.keys():
    classify(img_file)
    print('\n')
```
